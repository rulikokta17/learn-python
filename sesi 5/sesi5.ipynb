```python
# Bubble Sort
"""
Docstring
Algoritma pseudocode
bubble sort:
array awal : [5, 3, 8, 4, 2]
mulai
pjglist = len(array)
untuk i dari 0 ke pjglist-1
    untuk j dari 0 ke pjglist-i-1:
           if arr[j] > arr[j+1]:
                tukar arr[j] dan arr[j + 1]
selesai
"""

# buat function dari bubble sort (dapat digunakan berulang kali)
# Indentasi (spasi)
def bubble_sort(arr):
    # mulai
    pjg_list = len(arr) # kita ambil panjang list
    # lakukan perulangan sepanjang list
    for i in range(0, pjg_list-1):
        # lakukan looping dalam looping untuk membandingkan 2 data
        for j in range(0, pjg_list - i - 1):
            # pengecekan 2 data bersebelahan
            if arr[j] > arr[j + 1]: # apabila true
                # tuker datanya/posisinya
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# cara penggunaan function

list_gaji = [2000, 1000, 500, 3400, 5000]
print("Data sebelum diurutkan: ", list_gaji)
print("Data setelah diurutkan: ", bubble_sort(list_gaji))
```

```python
# Aplikasi Pengelolaan Nilai Siswa
"""
Fitur aplikasi
1. Melihat Nilai Siswa
2. Mengurutkan Nilai Siswa
3. Mengurutkan Nama Siswa
4. Keluar
"""

# dictionary -> mapping
siswa = [ # tipe data [] -> list
    {"nama": "Sigit", "nilai": 80}, #tipe data dict
    {"nama": "Angela", "nilai": 69},
    {"nama": "Patt", "nilai": 90},
    {"nama": "Rulik", "nilai": 70},
    {"nama": "Farel", "nilai": 88},
]

# buat function untuk tampilkan nilai
def tampilkan_nilai(siswa):
    print("\nDaftar Nilai Siswa:")
    for s in siswa:
        print(f"- {s['nama']}: {s['nilai']}")

# function bubble sort untuk nilai
def bubble_sort_by_nilai(siswa):
    pjg_data = len(siswa)
    for i in range (0, pjg_data - 1):
        for j in range (0, pjg_data - i - 1):
            if siswa[j]['nilai'] > siswa[j + 1]['nilai']:
                siswa[j], siswa[j + 1] = siswa[j + 1], siswa[j]
    return siswa      

# function bubble sort untuk nama
def bubble_sort_by_nama(siswa):
    pjg_data = len(siswa)
    for i in range (0, pjg_data - 1):
        for j in range (0, pjg_data - i - 1):
            if siswa[j]['nama'] > siswa[j + 1]['nama']:
                siswa[j], siswa[j + 1] = siswa[j + 1], siswa[j]
    return siswa  

# menu utamanya
def main():
    
    while True:
        print("\n===Aplikasi Pengelolaan Nilai Siswa===")
        print("1. Lihat Daftar Nilai SIswa")
        print("2. Urutkan Berdasarkan Nilai Siswa")
        print("3. Urutkan Berdasarkan Nama Siswa")
        print("4. Tambah Nilai Siswa")
        print("5. Keluar")
        # variable pilihan user
        pilihan = input("Masukkan Pilihan (1 - 5): ")
        
        # cek pilihan dari user
        if pilihan == "1":
            # panggil function tampilkan nilai siswa
            tampilkan_nilai(siswa)
        elif pilihan == "2":
            # panggil function sort by nilai
            hasil = bubble_sort_by_nilai(siswa)
            tampilkan_nilai(hasil)
        elif pilihan == "3":
            hasil = bubble_sort_by_nama(siswa)
            tampilkan_nilai(hasil)
        elif pilihan == "4":
            nama = input("Masukkan Nama Siswa: ")
            nilai = int(input("Masukkan Nilai Siswa: "))
            siswa.append({'nama': nama, 'nilai': nilai})
            print(f"Nilai siswa {nama} berhasil ditambah")
            
        elif pilihan == "5":
            print("Terima kasih sudah menggunakan Aplikasi ini")
            break
        else:
            print("Pilihan Tidak Valid")
            
# panggil main function
main()    
```
